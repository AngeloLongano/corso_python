#### Funzione Range
La funzione range è un generatore. Restituisce una sequenza di numeri in base ai parametri
-  range(n) --> fino ad n
-  range(min,max) --> da min a max
-  range(min,max,step) --> da min a max, con step (con step di 2, allora la seq. diventa 0,2,4,6,...)

### Iteratori e iterabili
In Python qualunque oggetto in grado di essere trattato come una sequenza è definito un oggetto iterable.
Un iterabili si suddivide in iteratori.
oggetto successivo: iteratore.next()

Note: usare i generatori (come iterabili) è molto più efficiente di creare delle liste

```python
#Esempio di iteratore infinito
class Potenza:
    def __init__(self, base):
        self.base = base
    def __iter__(self):
        self.numero = self.base
        return self
    def __next__(self):
        self.numero_successivo = self.numero
        self.numero = self.numero * self.base
        return self.numero_successivo
```

### Passaggio di parametri

-   fun(1,3,2) --> posizionali ( *args)
-   fun(a=1,c=2,b=3) --> con chiavi ( **kwargs)
-   fun(1,c=2,b=3) --> mix

##### * e ** servono per spacchettare:
-   \* serve per liste o tuple --> complex(3, 5) ↔ complex(*[3, 5])
-   ** serve per dizionari --> complex(real=3, imag=5) ↔ complex(**{'real': 3, 'imag': 5}) ---> (3+5j)
